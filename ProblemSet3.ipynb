```python
#Step 1
import pandas as pd
```

```python
#Step 2 
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
```

```python
#Step 3 
users = pd.read_csv(url,delimiter='|')
```

```python
#Step 4
mean_age_by_occupation = users.groupby('occupation')['age'].mean()
```

```python
#Step 5
users['male'] = (users['gender'] == 'M').astype(int)
male_ratio_per_occupation = users.groupby('occupation')['male'].mean().sort_values(ascending=False)
```

```python
#Step 6
age_range_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
```

```python
#Step 7
mean_age_by_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()
```

```python
#Step 8
total_people_by_occupation = users.groupby('occupation')['gender'].count()
percentage_women_per_occupation = (users.groupby('occupation')['gender'].value_counts() / total_people_by_occupation) * 100
```

```python
print("Mean Age per Occupation:\n", mean_age_by_occupation)
print("\nMale Ratio per Occupation (sorted):\n", male_ratio_per_occupation)
print("\nAge Range per Occupation:\n", age_range_per_occupation)
print("\nMean Age per Occupation and Sex:\n", mean_age_by_occupation_sex)
print("\nPercentage of Women and Men per Occupation:\n", percentage_women_per_occupation)
```

```python
#Quetion 5
# Step 1
import pandas as pd

# Step 2
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

# Step 3
chipo = pd.read_csv(url, sep='\t')



# Step 4
print(chipo.head(10))

# Step 5
num_observations = chipo.shape[0]
print("Number of observations:", num_observations)

# Step 6
num_columns = chipo.shape[1]
print("Number of columns:", num_columns)

# Step 7
print("Column names:", chipo.columns)

# Step 8
print("Index type:", chipo.index)

# Step 9
most_ordered_item = chipo['item_name'].value_counts().idxmax()
print("Most-ordered item:", most_ordered_item)

# Step 10
items_ordered = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
print("Number of items ordered:", items_ordered)

# Step 11
most_ordered_choice = chipo['choice_description'].value_counts().idxmax()
print("Most-ordered item in choice_description:", most_ordered_choice)

# Step 12
total_items_ordered = chipo['quantity'].sum()
print("Total items ordered:", total_items_ordered)

# Step 13
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
print("Item price type after conversion:", chipo['item_price'].dtype)

# Step 14
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("Revenue for the period:", revenue)

# Step 15
num_orders = chipo['order_id'].nunique()
print("Number of orders:", num_orders)

# Step 16
average_revenue_per_order = revenue / num_orders
print("Average revenue per order:", average_revenue_per_order)

# Step 17
num_different_items = chipo['item_name'].nunique()
print("Number of different items sold:", num_different_items)
```

```python
#Question 6
import numpy as np
import matplotlib.pyplot as plt

# Generate example years
years = np.arange(1867, 2015)

# Generate example data for marriages and divorces
marriages_per_capita = np.random.uniform(1, 10, size=len(years))
divorces_per_capita = np.random.uniform(0.5, 8, size=len(years))

# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(years, marriages_per_capita, label='Marriages per Capita', marker='o')
plt.plot(years, divorces_per_capita, label='Divorces per Capita', marker='s')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number per Capita')
plt.title('Example: Marriages and Divorces per Capita in the U.S. (1867-2014)')

# Show legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
#Question 7
import numpy as np
import matplotlib.pyplot as plt

# Years of interest
years = [1900, 1950, 2000]

# Generate example data for marriages and divorces per capita
marriages_per_capita = [np.random.uniform(1, 10), np.random.uniform(1, 10), np.random.uniform(1, 10)]
divorces_per_capita = [np.random.uniform(0.5, 8), np.random.uniform(0.5, 8), np.random.uniform(0.5, 8)]

# Bar width
bar_width = 0.35

# Create the bar chart
fig, ax = plt.subplots()
marriages_bars = ax.bar(years, marriages_per_capita, bar_width, label='Marriages per Capita')
divorces_bars = ax.bar(np.array(years) + bar_width, divorces_per_capita, bar_width, label='Divorces per Capita')

# Add labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Number per Capita')
ax.set_title('Example: Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
ax.set_xticks(np.array(years) + bar_width / 2)
ax.set_xticklabels(years)
ax.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
#Question 8
import numpy as np
import matplotlib.pyplot as plt

# Actors and their corresponding kill counts (example data)
actors = ['Matthew Mccaunehey', 'Leonardo Dicaprio', 'Brad Pitt', 'Robert DeNiro']
kill_counts = [56, 42, 71, 29]

# Sort actors and kill counts by descending order
sorted_indices = np.argsort(kill_counts)[::-1]
sorted_actors = [actors[i] for i in sorted_indices]
sorted_kill_counts = [kill_counts[i] for i in sorted_indices]

# Create the horizontal bar chart
fig, ax = plt.subplots()
bars = ax.barh(sorted_actors, sorted_kill_counts, color='skyblue')

# Add labels and title
ax.set_xlabel('Kill Count')
ax.set_ylabel('Actor')
ax.set_title('Deadliest Actors in Hollywood')

# Add the corresponding actor's name above each bar
for bar in bars:
    ax.text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
            str(int(bar.get_width())), va='center', ha='left')

# Show the plot
plt.tight_layout()
plt.show()
```
