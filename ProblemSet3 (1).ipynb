```python
#Step 1
import pandas as pd
```

```python
#Step 2 
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
```

```python
#Step 3 
users = pd.read_csv(url,delimiter='|')
```

```python
#Step 4
mean_age_by_occupation = users.groupby('occupation')['age'].mean()
```

```python
#Step 5
users['male'] = (users['gender'] == 'M').astype(int)
male_ratio_per_occupation = users.groupby('occupation')['male'].mean().sort_values(ascending=False)
```

```python
#Step 6
age_range_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
```

```python
#Step 7
mean_age_by_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()
```

```python
#Step 8
total_people_by_occupation = users.groupby('occupation')['gender'].count()
percentage_women_per_occupation = (users.groupby('occupation')['gender'].value_counts() / total_people_by_occupation) * 100
```

```python
print("Mean Age per Occupation:\n", mean_age_by_occupation)
print("\nMale Ratio per Occupation (sorted):\n", male_ratio_per_occupation)
print("\nAge Range per Occupation:\n", age_range_per_occupation)
print("\nMean Age per Occupation and Sex:\n", mean_age_by_occupation_sex)
print("\nPercentage of Women and Men per Occupation:\n", percentage_women_per_occupation)
```

```python
#Step 1
import pandas as pd

# Step 2
url = "https://gc.blackboard.com/bbcswebdav/pid-9084226-dt-content-rid-63597837_1/xid-63597837_1"

data = pd.read_csv(url, sep='\t')

#Step3
data.set_index('Yr_Mo_Dy', inplace=True)

# Step 4
def fix_year(year):
    if year >= 61:
        return 1900 + year
    else:
        return 2000 + year
    
    
#Step 5
data.index.year = data.index.year.map(fix_year)

# Step 6
missing_values = data.isnull().sum()

# Step 7
non_missing_count = data.count().sum()

# Step 8
mean_windspeed = data.values.mean()

# Step 9
loc_stats = data.describe(percentiles=[])

# Step 10
day_stats = data.resample('D').apply(['min', 'max', 'mean', 'std'])

# Step 11
jan_avg = data[data.index.month == 1].mean()

# Step 12
yearly_data = data.resample('Y').mean()

# Step 13
monthly_data = data.resample('M').mean()

# Step 14
weekly_data = data.resample('W').mean()

# Step 15
weekly_stats = weekly_data.loc['1961-01-02':'1961-12-31'].agg(['min', 'max', 'mean', 'std'], axis=0)

# Print the results
print("Missing values:\n", missing_values)
print("\nNon-missing count:", non_missing_count)
print("\nMean windspeed:", mean_windspeed)
print("\nLocation statistics:\n", loc_stats)
print("\nDay statistics:\n", day_stats)
print("\nAverage windspeed in January:\n", jan_avg)
print("\nYearly data:\n", yearly_data)
print("\nMonthly data:\n", monthly_data)
print("\nWeekly data:\n", weekly_data)
print("\nStatistics for the first 52 weeks:\n", weekly_stats)
```

```python
#Quetion 5
# Step 1
import pandas as pd

# Step 2
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

# Step 3
chipo = pd.read_csv(url, sep='\t')



# Step 4
print(chipo.head(10))

# Step 5
num_observations = chipo.shape[0]
print("Number of observations:", num_observations)

# Step 6
num_columns = chipo.shape[1]
print("Number of columns:", num_columns)

# Step 7
print("Column names:", chipo.columns)

# Step 8
print("Index type:", chipo.index)

# Step 9
most_ordered_item = chipo['item_name'].value_counts().idxmax()
print("Most-ordered item:", most_ordered_item)

# Step 10
items_ordered = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
print("Number of items ordered:", items_ordered)

# Step 11
most_ordered_choice = chipo['choice_description'].value_counts().idxmax()
print("Most-ordered item in choice_description:", most_ordered_choice)

# Step 12
total_items_ordered = chipo['quantity'].sum()
print("Total items ordered:", total_items_ordered)

# Step 13
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
print("Item price type after conversion:", chipo['item_price'].dtype)

# Step 14
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("Revenue for the period:", revenue)

# Step 15
num_orders = chipo['order_id'].nunique()
print("Number of orders:", num_orders)

# Step 16
average_revenue_per_order = revenue / num_orders
print("Average revenue per order:", average_revenue_per_order)

# Step 17
num_different_items = chipo['item_name'].nunique()
print("Number of different items sold:", num_different_items)
```

```python
#Question 7
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r'C:\Users\sirab\Downloads\us-marriages-divorces-1867-2014.csv')

# Extract data for the years 1900, 1950, and 2000
years = [1900, 1950, 2000]
marriages_per_capita = df[df['Year'].isin(years)]['Marriages_per_1000']
divorces_per_capita = df[df['Year'].isin(years)]['Divorces_per_1000']

# Create the vertical bar chart
plt.figure(figsize=(8, 6))
plt.bar(years, marriages_per_capita, width=0.4, align='center', label='Marriages per Capita')
plt.bar([year + 0.4 for year in years], divorces_per_capita, width=0.4, align='center', label='Divorces per Capita')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number per 1000')
plt.title('Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')

# Add legend
plt.legend()

# Adjust x-axis ticks
plt.xticks([year + 0.2 for year in years], years)

# Display the plot
plt.show()
```

```python
#Question 8
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r'C:\Users\sirab\Downloads\actor_kill_counts.csv')


# Sort by kill count
df_sorted = df.sort_values(by='Count', ascending=False)

# Select the top actors
top_actors = df_sorted.head(10)

# Create the horizontal bar chart
plt.figure(figsize=(10, 6))

plt.barh(top_actors['Actor'], top_actors['Count'], color='skyblue')
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')
plt.gca().invert_yaxis()
plt.tight_layout()

# Add labels to the bars
for index, value in enumerate(top_actors['Count']):
    plt.text(value, index, str(value))

plt.show()
```

```python
#Question 9
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data_path = r'C:\Users\sirab\Downloads\roman-emperor-reigns.csv'
df = pd.read_csv(data_path)

# Count the number of assassinated emperors
assassinated_count = df[df['Cause_of_Death'] == 'Assassinated']['Emperor'].count()

# Count the number of total emperors
total_emperors = df['Emperor'].count()

# Calculate the fraction of assassinated emperors
assassinated_fraction = assassinated_count / total_emperors

# Create a pie chart
labels = ['Assassinated', 'Not Assassinated']
sizes = [assassinated_fraction, 1 - assassinated_fraction]
colors = ['red', 'green']
explode = (0.1, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add title
plt.title('Fraction of Roman Emperors Assassinated')

# Display the pie chart
plt.show()

```
